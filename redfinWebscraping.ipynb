get listings of houses sold in last year

```python
#houses/condos/townhouses sold in last year in berkeley
url_sold = 'https://www.redfin.com/city/1590/CA/Berkeley/filter/sort=hi-sale-date,property-type=house,include=sold-1yr'
response = requests.get(url_sold, headers = header)

encod = response.encoding
contents = response.content.decode(encod)
soup = BeautifulSoup(contents, 'html.parser')

num_pages = soup.find_all('a',{'class':'clickable goToPage'}) #gets number of pages
num_pages = (len(num_pages))
```

```python
cnt = 0
houses_sold = []
address_lst = []
while cnt <= num_pages:
        url_temp = url_sold + '/page-' + str((cnt + 1)) #link for each page

        response = requests.get(url_temp, headers = header) #requests for html data
        encod = response.encoding
        contents = response.content.decode(encod)

        soup = BeautifulSoup(contents, 'html.parser')

        test = soup.find_all("div", {"class": "bottomV2"}) #gets all house cards
        for a in test:
                other = a.find_all('a', href = True) #looks for link
                string_link = str(other)
                temp = string_link[10:] #cuts off first part
                cut = left_text = temp.partition("\"")[0] #partitions the section we want of link
                cut = "https://www.redfin.com" + cut
                houses_sold.append(cut)
                
                address = string_link.partition("title=\"")[2]
                address = address.partition("\"")[0]
                address_lst.append(address)
        cnt += 1
print(houses_sold)
print(address_lst)
```

```python
df_sold = pd.DataFrame()
i = 0
for link in houses_sold:
    response = requests.get(link, headers = header)
    encod = response.encoding
    contents = response.content.decode(encod)
    soup = BeautifulSoup(contents, 'html.parser')

    #address = soup.find('header', {'class':'address'}).text
    try:
     address = address_lst[i]
     sqft = soup.find('span',{'class':'statsValue'}).text

     home = soup.find_all('div', {'class':'DPTableDisplay'})
     home = home[2]
     home_info = home.find_all('div', {'class':'table-value'})

     stories = home_info[3]
     lot_size = home_info[4]
     house_type = home_info[5]
     yr_built = home_info[6]
     

     info = soup.find_all('div', {'class':'statsValue'}) 
     price = info[0].text
     beds = info[1].text
     baths = info[2].text
    except:
       address = None
       sqft = None
       home = None
       stories = None
       lot_size = None
       house_type = None
       yr_built = None

    d = {'Beds':[beds], 'Baths':[baths], 'Square Ft':[sqft],'Type':house_type, 
         'Year Built':yr_built,'Lot Size':lot_size,'Stories':stories, 'Address':[address], 'Price':[price]}
    df_temp = pd.DataFrame(data = d)
    df_sold = pd.concat([df_temp, df_sold])
    i += 1
```
